## Part 3: Scraping Data from Wikipedia

I selected Wikipedia’s page about the most expensive association soccer transfers to web-scrape data.

This page is works well with this project because it focuses on soccer transfer fees and by looking at this data we can see another way how to undertand athlete market value.

```python
import requests
from io import StringIO

url = "https://en.wikipedia.org/wiki/List_of_most_expensive_association_football_transfers"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

tables = pd.read_html(StringIO(response.text))

len(tables)
```

```python
for i, table in enumerate(tables[:5]):
    print("Table number:", i)
    display(table.head())
```

```python
transfers = tables[0]

transfers.head()
```

```python
transfers.info()
```

```python
transfers.columns
```

```python
transfers.columns = [str(column).strip().replace("\n", " ") for column in transfers.columns]

transfers.head()
```

```python
soccer_transfers = transfers.copy()

soccer_transfers.head()
```

```python
soccer_transfers.columns
```

```python
soccer_transfers = soccer_transfers.rename(columns={
    "Player": "Player",
    "Position": "Position",
    "Year": "Year",
    "Fee (€ million)": "Fee_Euro_Million"
})

soccer_transfers.head()
```

```python
soccer_transfers.columns
```

```python
soccer_transfers["Fee_Euro_Million"] = (
    soccer_transfers["Fee_Euro_Million"]
    .astype(str)
    .str.extract(r"(\d+\.?\d*)")[0]
    .astype(float)
)

soccer_transfers.head()
```

```python
soccer_transfers[["Player", "Position", "Fee_Euro_Million", "Year"]].head(10)
```

```python
soccer_transfers.isna().sum()
```

## Wikipedia Data Cleaning

## Part 4: Visualizing and Analyzing the Data

The first visualization uses the ISLP `Hitters` dataset. It compares player experience with salary.

```python
hitters_clean = hitters.dropna(subset=["Salary"])

hitters_clean.shape
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(hitters_clean["Years"], hitters_clean["Salary"])
plt.xlabel("Years Played")
plt.ylabel("Salary in Thousands of Dollars")
plt.title("Player Salary by Years Played")
plt.show()
```

## Analysis of Visualization 1

By looking at the scatter plot that I created we can see that players that are more ecperience in years, earn higher salaries. However, the relationship between years played and salary is not perfect. The reason why is by looking at the scatter plott we can see that players with less years of experience could still earn high saleries and players with many years of experience could earn lower salaries. To summarize, this scatter plot explais that their are other factors that could affect the players salary, and not just because of their experience in years. 

## Second Visualization 

The second visualization uses the Wikipedia soccer transfer data. It shows the average transfer fee by player position.

```python
position_fees = (
    soccer_transfers
    .groupby("Position")["Fee_Euro_Million"]
    .mean()
    .sort_values(ascending=False)
)

position_fees
```

```python
plt.figure(figsize=(8, 5))
position_fees.plot(kind="bar")
plt.xlabel("Position")
plt.ylabel("Average Transfer Fee in € Millions")
plt.title("Average Soccer Transfer Fee by Position")
plt.xticks(rotation=45)
plt.show()
```

## Analysis of Visualization 2

By looking at the bar chart we can see that depending of what position tha player is playing they have higher or lower market value in elite soccer transfers. For example, a defender might not be able to as high transfer fees as a forward, because a forward might be axtra valued by the club if they score more goals. In conlcusion, this data-set from Wikipedia only contains the most expensive transfers, which means the the entire soccer market its not show in this data-set.
